```python
# from https://stackoverflow.com/questions/30328646/python-pandas-group-by-in-group-by-and-average
import numpy as np
import pandas as pd
df = pd.DataFrame()
```

```python
# df = pd.DataFrame(columns = ['cluster','org','time'])
df = pd.DataFrame(columns=["cluster", "org", "time"], 
                  data=[[1,'a',8],
                       [1,'a',6],
                       [2,'h',34],
                       [1,'c',23],
                       [2,'d',74],
                       [3,'w',6]])
```

```python
df
```

```python
df.groupby(['cluster']).mean()
```

```python
(df.groupby(['cluster', 'org'], as_index=False).mean()
            .groupby('cluster')['time'].mean())
```
